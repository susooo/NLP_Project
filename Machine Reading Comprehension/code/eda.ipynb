```python
MODEL_NAME = "monologg/kobigbird-bert-base" #"monologg/koelectra-base-v3-discriminator"
dataset = load_dataset('custom_squad_v2', split='train')
# squad = dataset.train_test_split(0.1, seed=SEED)
```

```python
anwsers = dataset['answers']

max_ = 0
sum_ = 0
lengths = []

for answer in anwsers:
    for text in answer['text']:
        #@print(text)
        l = len(text)
        lengths.append(l)
        
        if l > max_:
            max_ = l
        
        sum_ += l

print(f'Mean: {sum_/len(lengths)}, Max: {max_}')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.boxplot(lengths)

ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# 3. 그래프 그리기
fig, ax = plt.subplots()

box = ax.boxplot(lengths, notch=True, whis=2)
ax.set_ylim(0, 25.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

whiskers = [item.get_ydata() for item in box['whiskers']]
medians = [item.get_ydata() for item in box['medians']]
fliers = [item.get_ydata() for item in box['fliers']]
print(f'fliers: {len(fliers[0])}')
print('whiskers:', whiskers)
print('medians:', medians)
print('fliers:', fliers)

plt.show()
```
